```python
from transformers import AutoFeatureExtractor, AutoModelForAudioClassification
import torch
import torch.nn.functional as F
from transformers import pipeline
import soundfile as sf
```

```python
# Use a pipeline as a high-level helper
from transformers import pipeline
import scipy.io.wavfile as wav
from scipy.signal import resample
import numpy as np
```

```python
audio_classif=pipeline("audio-classification", model="VaidehiBhat007/fine_tuned_emo_classif")

speech2text= pipeline("automatic-speech-recognition", model="openai/whisper-large-v3")

llm= pipeline("text2text-generation", model="facebook/blenderbot-400M-distill")

text2speech=pipeline("text-to-speech", "microsoft/speecht5_tts")
```

```python
def mainfn(audio):
    try:
        # Read the audio file
        sampling_rate, data = wav.read(audio)
        print("Original sampling rate:", sampling_rate)

        # Resample the audio to 16000 Hz
        new_sampling_rate = 16000
        num_samples = int(len(data) * new_sampling_rate / sampling_rate)
        resampled_data = resample(data, num_samples)
        print("Resampled data shape:", resampled_data.shape)

        # Save the resampled audio to a temporary file
        temp_audio_path = "resampled_audio.wav"
        wav.write(temp_audio_path, new_sampling_rate, np.asarray(resampled_data, dtype=np.int16))
        print("Saved resampled audio to:", temp_audio_path)

        # Load the resampled audio for classification
        audio_input, _ = sf.read(temp_audio_path)
        outputs = audio_classif(temp_audio_path)
        print("Audio classification outputs:", outputs)

        # Extract predictions and determine the emotion
        predictions = torch.nn.functional.softmax(torch.tensor([output['score'] for output in outputs]), dim=-1)
        predicted_label = predictions.argmax().item()
        print("Predicted label:", predicted_label)

        # Map the predicted label to an emotion
        emotion_labels = ["neutral", "happy", "disgusted", "fearful", "sad", "angry"]
        if predicted_label < len(emotion_labels):
            predicted_emotion = emotion_labels[predicted_label]
        else:
            return "No emotion detected", None
        print("Predicted emotion:", predicted_emotion)

        # Convert speech to text
        text_result = speech2text(temp_audio_path)
        text = text_result["text"]
        print("Transcribed text:", text)

        # Generate a response based on the text and emotion
        prompt = (
            "You are a conversational assistant. You need to keep the conversation going. "
            "Given a text and an emotion associated with it, your task is to reply to the text. "
            "Your tone should be suitable as a reply for the user's emotion. Clearly mention the emotion given. "
            f"Text: {text} Emotion: {predicted_emotion}"
        )
        result2 = llm(prompt)
        to_audio = result2[0]['generated_text']
        print("Generated response:", to_audio)

        # Convert the generated text back to speech
        speech = text2speech(to_audio)
        print("Generated speech audio data length:", len(speech["audio"]))

        # Save the generated speech to a file and return it
        output_audio_path = "speech.wav"
        sf.write(output_audio_path, speech["audio"], samplerate=speech["sampling_rate"])
        print("Saved generated speech to:", output_audio_path)

        return output_audio_path

    except Exception as e:
        print("Error:", e)
        return str(e), None
```

```python
iface = gr.Interface(
    fn=mainfn,
    inputs=gr.Audio(sources=["microphone"], type="filepath", label="Speak Here!"),
    outputs=gr.Audio(type="filepath", label="Here's What I got to say"),
    title="Your personal Audio Chatbot!",
    description="Record to go!"
)

iface.launch()
```
